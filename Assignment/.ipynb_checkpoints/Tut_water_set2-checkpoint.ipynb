Fill in the blanks each 2 marks

```python
def linear(X, b0, b1):
    return [b0+b1*x for x in X]
```

```python
# b0 - Intercept
def intercept(X, Y, b1): 
    x_ = np.mean(X)
    y_ = np.mean(Y)
    
    return y_- b1*x_ #2 marks
```

```python
# b1 - Slope
def slope(X, Y):
    x_ = np.mean(X)
    y_ = np.mean(Y)
    
    sxy = sum([(x-x_)*(y-y_) for x,y in zip(X,Y)])
    sxsq = sum([(x-x_)**2 for x in X])
    
    return sxy / sxsq
```

```python
data = pd.read_csv("Nuclear_Energy.csv")
data.head()
```

```python
#take year attribute in predictor and Belgium attribute in target----1 mark
yr = data['Years']
target = data['Belgium (Nuclear % share in power gen)']
```

```python
b1 = slope(yr, target)
b0 = intercept(yr, target, b1)
predicted = linear(yr, b0, b1)
```

```python
#Calculate RMSE value and display---2 marks
rmse = np.sqrt(sum([(y-y_)**2 for y,y_ in zip(target, predicted)])/len(target))
print(rmse)
```

```python
plt.figure(figsize = (10, 5))
#scatter plot the input data
plt.scatter(yr, target, color = 'purple', label = "Actual Data")
plt.title('Predicted values by Linear Regression', fontsize = 22)
plt.xlabel('year')
plt.ylabel('BelgiumNP')
# plot the predicted data-----1 mark
plt.plot(yr, predicted, color = 'green',marker=".",markersize="10", label = 'Predicted Line')
plt.legend()
plt.show()
```
