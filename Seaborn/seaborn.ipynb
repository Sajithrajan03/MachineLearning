```python
import seaborn as sns

df = sns.load_dataset("iris")
df
```

```python
plt = sns.lineplot(x="sepal_length",y="sepal_width",data = df)
```

```python
palette = sns.color_palette()
sns.palplot(palette)
plt.show()
```

```python
sns.stripplot(x="sepal_length",y="sepal_width",data = df)
```

```python
sns.histplot(x="sepal_length",y="sepal_width",data = df)
```

```python
sns.distplot(df["sepal_length"])
```

```python
sns.pairplot(data = df,hue='species')
```
